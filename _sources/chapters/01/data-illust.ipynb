# Data Illustration

## SEC-SAXS Experiments

The data we will discuss here come from experiments conducted in device systems such as

* [SEC-SAXS System in SPring-8, Japan](https://new.spring8.or.jp/index.php/home/2021-03-23-05-27-47/2-uncategorised/327-sec-saxs).[^1]

Acknowledging importance of the hardware acpects, we set them aside and start from a situation where data are given, duly processed as input to `Molass Library`.[^2]

That said, to share the discussion context, we need a minimum idea of what they are like, which will be illustrated below from some different points of view.

[^1]: Or, see also [an excellent video](https://www.youtube.com/watch?v=aoth3Fq7DCE) by [BioCat](https://www.youtube.com/@biocat_iit) on YouTube.

[^2]: The presice requirements for the input data should be defined elsewhere.

## 3D View

Since the data are measured in two different ways, namely by UV absorbance and X-ray scattering, they can be expressed as a whole in 3D plots as follows.


```{figure} ../../_static/images/corrected_ssd_3d.png
:name: three_d_view_of_the_data

3D view of SEC-SAXS data
```

```{note}
We say here, for lack of a better word, "spectral" to suggest the same direction for both of absorbance and scattering data. For the latter, "angular" might be used in other contexts since the Q-axis represents the scattering angle, or presicely, momentum transfer $ q = 4 \pi \sin(\theta)/\lambda $ where $ 2\theta $ is the scattering angle and $\lambda$ the incident X-ray wavelength.
```
