## Mathematical Relation between an Object and its Scattering Curve

In a real experiment, scaterred X-ray intensities are measured using a detector,
which produces data to be circular-averaged into a scaterring curve.

Mathematically, these are related by a Fourier transform between an electron density space
and its reciprocal space whose spherical average correspond to the scattering curve.

This relation is depicted below using SAXS simulator in the Molass Library.

We take this relation for granted in this book and will not give details any further.

See {cite:p}`FeiginSvergun1987`, {cite:p}`Svergun2013` or {cite:p}`Lattman2018` for details.
