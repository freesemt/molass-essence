# SAXS Basics

## Mathematical Relation between Particles and its Scattering Curve

In a real experiment, X-rays are radiated at particles
and scaterred X-ray intensities are measured using a detector,
which produces data to be circular-averaged into a scaterring curve.

Mathematically, this expriment is expressed by some data sets
which are related by a Fourier transform between an electron density space and its reciprocal space,
where the spherical average of reciprocal space corresponds to the scattering curve.

This relation is depicted below using SAXS simulator in the Molass Library.

```{note}
We take this relation for granted in this book and will not give details any further,
although we will show how to use this relation to understand SAXS.
See {cite:p}`FeiginSvergun1987`, {cite:p}`Svergun2013` or {cite:p}`Lattman2018` for details.
```
